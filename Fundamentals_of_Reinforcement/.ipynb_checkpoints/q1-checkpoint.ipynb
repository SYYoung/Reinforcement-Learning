```python
a = np.array([1,2,3])
np.insert(a, len(a), 20)
```

```python
def plt_target_est(data1, data2):
    plt.plot(data1)
    plt.plot(data2)
    plt.show()
```

```python
def estimate(alpha):
    new_e = np.zeros(len(target)+1)
    for i in range(len(target)):
        r = target[i]
        new_e[i+1] = new_e[i] + alpha[i] * (r - new_e[i])
    return new_e
```

```python
a1 = np.array([1.0 for i in range(len(target))])
e1_0 = estimate(a1)
```

```python
plt_target_est(target, e1_0)
```

```python
plt_target_est(target, estimate(0.5))
```

```python
plt_target_est(target, estimate(0.125))
```

```python
a1 = np.array([1/i if i > 0 else 1 for i in range(len(target))])
print(a1)
e1_0 = estimate(a1)
plt_target_est(target, e1_0)
```
